# Exploring distributions of numeric columns using histograms

```python
import warnings

# supree futer warning about  use_inf_as_na
warnings.simplefilter(action='ignore', category=FutureWarning)
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

df = pd.read_csv("C:/Users/prady/Downloads/WineQT.csv")

# set the style for sea born
sns.set(style="whitegrid")

#Create sub plot for each neumeric column
fig, axes= plt.subplots(nrows=3, ncols=4, figsize=(15, 10))

# Flatten the axes array for easire indexing
axes = axes.flatten()

# List of neumeric column
numeric_columns = df.select_dtypes(include=['float64']).columns

# Create histogram for each neumeric columns 
for i, column in enumerate(numeric_columns):
    sns.histplot(df[column], ax=axes[i], kde=True, bins=20, color='red')
    axes[i].set_title(f'Distribution of {column}')

# Adjust layout
plt.tight_layout()
plt.show()
```

# Four Interesting question about the dataset

### 1)What is the distribution of 'fixed acidity' in the dataset?



```python
import seaborn as sns
import matplotlib.pyplot as plt

# Histogram for 'fixed acidity' distribution
plt.figure(figsize=(8, 6))
sns.histplot(df['fixed acidity'], bins=20, kde=True, color='blue', edgecolor='black')
plt.title('Distribution of Fixed Acidity')
plt.xlabel('Fixed Acidity')
plt.ylabel('Frequency')
plt.show()
```

### 2)Is there a correlation between 'density' and 'pH'?

```python
# Scatter plot for 'density' vs 'pH'
plt.figure(figsize=(10, 6))
sns.scatterplot(x='density', y='pH', data=df, color='purple')
plt.title('Relationship between Density and pH')
plt.xlabel('Density')
plt.ylabel('pH')
plt.show()
```

### 3)What is the most common range of 'total sulfur dioxide' in the dataset?

```python
# Boxplot for 'total sulfur dioxide'
plt.figure(figsize=(8, 6))
sns.boxplot(x=df['total sulfur dioxide'])
plt.title('Distribution of Total Sulfur Dioxide')
plt.xlabel('Total Sulfur Dioxide')
plt.show()
```

### 4)How do 'volatile acidity' and 'citric acid' vary together?



```python
# scatter plot for volatile acid and citric acid 
plt.figure(figsize=(10, 6))
sns.scatterplot(x='volatile acidity', y='citric acid', data=df, color='green')
plt.title('Relationship between volatile acidity and citric acid')
plt.xlabel('volatile acidity')
plt.ylabel('citric acid')
plt.show()
```

# Created new columns total_acidity and sulfur_ratio

```python
df['total_acidity'] = df['fixed acidity'] + df['volatile acidity']
df['sulfur_ratio'] = df['free sulfur dioxide'] / df['total sulfur dioxide']
print(df.head())
```

# Merge Multiple Datasets (Creating a Duplicate for Illustration)

```python
#Creating a duplicate Data Frame for illuestration purpose
df2 = df.copy()

# merging the two data frames using the index as the common column
merged_df = pd.merge(df, df2, left_index=True, right_index=True,how='inner')

# Displayed the merge data frame
print(merged_df.head())
```

## Grouping and Aggregation

```python
# Grouping by 'pH' and calculating mean and sum for 'density and 'sulphates'
grouped_df = df.groupby('pH').agg({'density': 'mean', 'sulphates': 'sum'}).reset_index()

#Display the grouped and aggregated DataFrame 
print(grouped_df.head())
```
